```r
wilcox.test(NLF_NS_conc$Conc_pslog2,ELF_NS_conc$Conc_pslog2) 
wilcox.test(NLF_NS_conc$Conc_pslog2,Sputum_NS_conc$Conc_pslog2)
wilcox.test(NLF_NS_conc$Conc_pslog2,Serum_NS_conc$Conc_pslog2)
wilcox.test(ELF_NS_conc$Conc_pslog2,Sputum_NS_conc$Conc_pslog2)
wilcox.test(ELF_NS_conc$Conc_pslog2,Serum_NS_conc$Conc_pslog2)
wilcox.test(Sputum_NS_conc$Conc_pslog2,Serum_NS_conc$Conc_pslog2)
```
